```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
from scipy import ndarray
import skimage as sk
from skimage import transform
from skimage import util
import random
```

```python
path = os.path.join(os.getcwd(), 'newImage')
print(path)

categories = os.listdir(path)

for category in categories:
    file_folder_path = os.path.join(path, category)
    file_name = os.listdir(file_folder_path)
    for file in file_name:
        file_path = os.path.join(file_folder_path, file)
        img = mpimg.imread(file_path)
        img_reverse = img[:,::-1]
        img_noise = sk.util.random_noise(img)
        img_reverse_noise = sk.util.random_noise(img_reverse)
        
        num = len(os.listdir(file_folder_path))
        path_temp = os.path.join(file_folder_path, category + str(num+1) + '.jpg')
        sk.io.imsave(path_temp, img_reverse)
        
        num = len(os.listdir(file_folder_path))
        path_temp = os.path.join(file_folder_path, category + str(num+1) + '.jpg')
        sk.io.imsave(path_temp, img_noise)
        
        num = len(os.listdir(file_folder_path))
        path_temp = os.path.join(file_folder_path, category + str(num+1) + '.jpg')
        sk.io.imsave(path_temp, img_reverse_noise)
    


```
